## Data Preprocessing and Cleaning

```python
df = pd.read_csv("medical_insurance.csv", keep_default_na=False)
df
```

```python
# What sort of values are we dealing with
for column in df.columns:
    distinct_values = df[column].unique()
    print(f"Distinct values in column '{column}': {distinct_values}")
```

```python
df['alcohol_freq'] = df['alcohol_freq'].replace('None', 'Never')
```

```python
# Filter age group due to inconsistences and practicality
df = df[(df['age'] >= 18) & (df['age'] <= 65)]
df
```

```python
# Drop irrelevant details and columns that might leak the target
df = df.drop(columns=["person_id", "annual_premium", "monthly_premium", "claims_count", "avg_claim_amount", "annual_medical_cost"], axis=1)
```

```python
df.to_csv("cleaned_medical_insurance.csv", index=False)
```

## EDA

### Distribution Plots

```python
def plot_hist(df, col):
    plt.hist(df[col])
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

actual_numeric_cols = []

for col in list(df.columns):
    if is_numeric_dtype(df[col]) and not df[col].isin([0, 1]).all():
        actual_numeric_cols.append(col)
        plot_hist(df, col)        
```

### Log transform columns with heavy skew

```python
log_vars = [
    "income",
    "visits_last_year",
    "hospitalizations_last_3yrs",
    "days_hospitalized_last_3yrs",
    "medication_count",
    "proc_imaging_count",
    "proc_surgery_count",
    "proc_physio_count",
    "proc_consult_count",
    "proc_lab_count",
    "total_claims_paid"
]

for col in log_vars:
    df[f"log_{col}"] = np.log1p(df[col])

df
```

### Correlation

```python
actual_numeric_cols = []
for col in list(df.columns):
    if is_numeric_dtype(df[col]) and not df[col].isin([0, 1]).all():
        actual_numeric_cols.append(col)

new_list = [item for item in actual_numeric_cols if item not in log_vars]

correlation_matrix = df[new_list].corr()
sns.heatmap(correlation_matrix, cmap="coolwarm", center=0)
```

```python
# Filter the matrix to show only values > 0.7, excluding self-correlations and duplicates
# Create a boolean mask to keep only the upper triangle (excluding diagonal)
mask = np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
# Apply the mask and stack the results into a single Series
strong_corrs = correlation_matrix.where(mask).stack()
# Filter the stacked Series for values greater than 0.7
strong_corrs = strong_corrs[strong_corrs > 0.7]

strong_corrs
```

### Check for Multicollinearity (for MLR only)

```python
actual_numeric_cols = []
for col in list(df.columns):
    if is_numeric_dtype(df[col]) and col not in log_vars:
        actual_numeric_cols.append(col)
        
pre_vif = ColumnTransformer(
    [
        ("num", StandardScaler(), actual_numeric_cols),
    ]
)

pre_vif
```

```python
X_vif = pre_vif.fit_transform(df)

vif_data = pd.DataFrame()
vif_data["VIF"] = [variance_inflation_factor(X_vif, i) for i in range(X_vif.shape[1])]
vif_data["feature"] = pre_vif.get_feature_names_out()
vif_data
```

```python
vif_data[vif_data['VIF'] >= 10]
```
